```python
#import libraries
import numpy as np 
import pandas as pd 
print('imported')
```

Turn file to pandas dataframe

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df=pd.read_html(url, header=0)[0]
df.head(15)
```

Show all rows with 'not assigned' boroughs

```python
df3=df[df['Borough'] == 'Not assigned']
df3
```

Remove rows with 'not assigned' boroughs

```python
df1= df[df['Borough'] != 'Not assigned'] 
df1 = df1.reset_index(drop=True)
df1
```

Rows with 'not assigned' neightbourhood will have the borough as its neighborhood

```python
df1['Neighbourhood']=np.where(df1['Neighbourhood'] == 'Not assigned',df1['Borough'],df1['Neighbourhood'])
df1
```

Combine rows with same postal code

```python
df1=df1.groupby(['Postal Code','Borough'])['Neighbourhood'].agg(", ".join)
df1 = df1.reset_index()
df1
```

Sort rows in ascending order

```python
df1=df1.sort_values(by='Postal Code', ascending=True)
df1
```
